This has slightly more varied entries than the 4x3 matrix we used before, just so we don't get confused if we're accessing a location or a specific element if there is a "2" in location "2."

If you have an NxM matrix, MATLAB assigns a specific numeric location to each element between 1 and whatever the product of N and M is. (For example, in your 4x3 matrix, every element has a position 1 through 12.) **IF YOU HAVE PROGRAMMED IN OTHER LANGUAGES, YOU MAY BE EXPECTING THE MATRIX TO START AT ELEMENT 0. NOTE THAT IN MATLAB, THIS IS NOT THE CASE, AND ARRAYS START AT 1.**

We access these specific positions using the name of the matrix followed by parenthesis with the location. For example, if we named our matrix "m" and we want the eigth element, we can type `disp(m(8));` and this will display element 8 of the matrix.
