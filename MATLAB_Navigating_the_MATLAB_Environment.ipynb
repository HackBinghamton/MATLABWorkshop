## Understanding the Window

If this is your first time using MATLAB, there are probably some things you should know about the environment. 

On the left hand side, you have two boxes, one labelled "Current Folder," and the other labelled "Details."

"Current Folder" shows you where in your file system you are. When you first open MATLAB, you are automatically brought to your MATLAB folder. To run any of your code, you will need to navigate to the folder you have your file in. Since we currently have no folders, this doesn't affect us yet.

On the right hand side of the environment, there's a window titled "Workspace." We will be going over what this is later. For now, just make a mental note that it's empty.

In the center of the environment, there's a giant box labelled "Command Window." If you don't care about saving any code and just want to run one or two lines of code, you can do it here. Let's test it out. Type the following into your command window and then hit enter:

```x = 1 + 2;```

You should notice two things: first, a new line (denoted by the >>) popped up. This means that the line of code execuded successfully. Second, the "Workspace" window on the right hand side of the screen now has a new line.

2.PNG

The window has two columns, "Name" and "Value." MATLAB has saved both your variable's name and value in the workspace. If you want to use "X" again, the value "3" is stored in it.

Try entering the following:

```x + 1;```

### Why is the Workspace window Helpful?

Try double clicking on where it says "ans." A new window should pop up, "Variables," with a grid resembling an excel spreadsheet

3.PNG

This window has a tab labelled "ans," where it tells you it is a 1x1 double. A double is a way that MATLAB can save a number. 1x1 is referring to the dimension of what you saved.

MATLAB is mainly used when there will be a large amount of mathematics involved in a project or library. If in the future, you end up working with large matrices, this window can be extremely helpful in allowing you to view your variables without actually having to run your code again.

If you are working on large programs, it might take several minutes (or even several hours) for your code to run. You don't want to have to wait all that time to print out a variable to the console just to see if it's correct or not. MATLAB allows you to just click on and display any variable you used in the code you ran most recently.

This can also be helpful for when you are working on projects with lots of variables. Being able to look at all of them in alphabetical order on the side of your screen can help you stay organized and keep straight what you're working on.

Feel free to play around with this a bit more before moving on!


## Let's try coding!

Quit out of the "variables" tab. Now navigate to the tab where it says "New" with a giant yellow plus sign, and click "script." This should bring up a new window titled "Editor - Untitled." In this new window, type the following:

```disp("Hello World!");```

Now hit "Run," which whould be a green triangle towards the top of your screen. This will prompt you to save your code first. Save it as whatever filename you want (it is automatically saved as a .m file). 

4.PNG

Hopefully "Hello World!" was just printed to the command window. "disp()" is how we tell MATLAB to display something. (If you have coded in python previously, this is similar to the "print()" function.) You will also notice that on the left-hand side, under "Current Folder," your saved file should appear! If you want to make another script, it will also show up here, and you can navigate your files using this window.
